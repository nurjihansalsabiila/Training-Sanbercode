<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 

# Perbedaan Series dengan List dan dictionary
Series menampilkan index dan nilainya, index tidak hanya bertipe integer. sedangkan 
List menampilkan nilainya saja

Series bisa memiliki key yang duplikat, dan mampu slicing data. sedangkan 
Dictionary tidak bisa memiliki key yang sama, tidak mampu slicing data

# Perbedaan dataframe dan series
DataFrame berstruktur 2 dimensi yang memiliki index colum dan index baris
Series berstruktur 1 dimensi saja atau 1 index

```python
# code here
import pandas as pd
hari=['Senin', 'Selasa','Rabu','Kamis','Jumat','Sabtu','Minggu']
pd.Series(hari)
```
