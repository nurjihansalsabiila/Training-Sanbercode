````python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
df = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/diamonds.csv')
df
````

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

````python
fig, ax = plt.subplots(figsize=(13, 13))
palette="ch:r=-.2,d=.3_r"
#code here
x=df['carat']
y=df['price']
clarity=df['clarity']
clarity_sort=['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
depth=df['depth']

sns.despine(fig, left=True, bottom=True)
sns.scatterplot(x, y, hue=clarity, hue_order=clarity_sort, palette=palette, size=depth, sizes=(1,8), linewidth=0)
ax.set_xlabel('Carat')
ax.set_ylabel('Price')

plt.show()
````

````python
#code here
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)

with sns.axes_style('dark'):
    grid=sns.FacetGrid(df, hue='cut', col='cut', height=4)
grid.map(hexbin,'carat', 'price', extent=[0, 3.0, 0, 20000]);
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df = df.sample(n=300, random_state=123)
data=df[['carat','depth','table','price','cut']]
#code here
with sns.axes_style('ticks'):
    plot=sns.pairplot(data, hue='cut', 
                      hue_order=['Premium','Ideal','Very Good','Good','Fair'], 
                      markers=['+','o','s','D','x'], 
                      diag_kind='kde', height=3, diag_kws=dict(shade=True))
    
    for i,j in zip(*np.triu_indices_from(plot.axes, 1)):
        plot.axes[i,j].set_visible(False)

plt.show()
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
# code here
df = df.sample(n=300, random_state=123)
x=df['carat']
y=df['price']

import scipy.stats as stats
import warnings
warnings.filterwarnings('ignore')

sns.set(style='white')
sns.jointplot(x, y, kind='kde', height=7, space=0)

sns.set(style='ticks')
sns.jointplot(x, y, kind='reg', xlim=(0, 2.5), ylim=(0, 20000), height=7, stat_func=stats.pearsonr)
````
