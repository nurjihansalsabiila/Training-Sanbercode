Machine learning adalah komputer yang memiliki kemampuan melakukan belajar dari pengalaman terhadap tugasnya dimasa lalu, dan mengalami peningkatan kinerja.

data feature -> variable yang berperan dalam melakukan prediksi

data target -> variable hasil prediksi

Supervised Learning -> Memiliki column data label sehingga output jelas

Unsupervised Learning -> tidak memiliki data label sehingga output tidak jelas

Jenis Supervised : Klasifikasi dan Regressi
1. Klasifikasi -> outpunya berupa class label
2. Regresi -> outputnya berupa quantity yang continue

Hyperparameter -> Didefinisikan diawal pembuatan objek model, tidak ada hubungannya dengan model (pengoptisasian model)

Parameter -> Dihasilkan diakhir, berada didalam model

1. Pilih model/algorithm yang akan digunakan

3. Pisahkan antara data feature dan data target
4. Perintahkan model untuk mempelajari data dengan menggunakan method.fit()

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
# Code here
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree

model = DecisionTreeRegressor(criterion='squared_error', max_depth=4, max_features=None,
                              min_samples_leaf=5, min_samples_split=2, min_weight_fraction_leaf=0.0,
                              random_state=None, splitter='best')

x_train=x.reshape(-1,1)
y_train=y
x_test=x_new.reshape(-1,1)

model.fit(x_train, y_train)
y_test=model.predict(x_test)
y_predict=model.predict(x_train)

fig=plt.subplots(figsize=(12,8))
plt.scatter(x, y, label='Data Asli')
plt.scatter(x_train, y_predict, label='Prediksi terhadap data training', color='y')
plt.plot(x_test, y_test, label=' Prediksi Terhadap Data Baru', color='g')
plt.legend(bbox_to_anchor=(1.1, 1.2))
plt.show()
tree.plot_tree(model)
```
