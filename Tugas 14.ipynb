```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('austin_weather.csv')
df.head()
```

```python
df.info()
```

<h2>Soal 1: Visualisasi Scatter Plot Perbandingan Kuantitatif</h2>

Pada tugas kali ini kita akan mengamati nilai DewPointAvg (F) dengan mengamati nilai  HumidityAvg (%), TempAvg (F), dan WindAvg (MPG)

Perhatikan bahwa data kita tidaklah siap untuk di analisis, salah satunya tipe data dari DewPointAvg (F), HumidityAvg (%), dan WindAvg (MPG) adalah object, padahalnya data nya ber isi numeric. maka :
- Ubahlah tipe data tersebut menjadi tipe data float

Kemudian: 

- Kalian tidak akan dengan mudah mengubah tipe data tersebut karena column tersebut mempunyai nilai '-' yang dimana tidak bisa di ubah ke bentuk float, maka replace lah terlebih dahulu data yang bernilai '-' dengan nilai NaN, gunakan method .replace(). baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.replace.html
- Isi nilai nan dengan nilai sebelumnya di row tersebut. gunakan method .fillna() dengan argument method bernilai 'ffill', baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html
- Sekarang ubah tipe datanya dengan float, gunakan method .astype(), baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.astype.html

Setelah ini sebagian data siap untuk di jadikan bahan analisis. maka:

Buahlah visualisasi perbandingan kuantitatif scatter plot, sehingga menghasilkan gambar seperti dibawah:

ket: 

- colormap adalah 'coolwarm'
- berikat warna terhadap setiap data poin dengan nilai dari column TempAvgF
- berikan size terhadap setiap data poin dengan nilai dari column WindAvgMPH, kalikan dengan 20 agar size terlihat lebih besar


Berikan pendapat dari insight yang bisa di dapat dari visualisasi perbandingan kuantitatif ini!!!

![alt text](https://drive.google.com/uc?id=1Pp5oIc-39mcw2vrw6xgCRM9GsvyRtPI5)

```python
# code here
df1=df.replace('-',np.NaN)
df2=df1.fillna(method='ffill')
df[['DewPointAvgF','HumidityAvgPercent','WindAvgMPH']]=df2[['DewPointAvgF','HumidityAvgPercent','WindAvgMPH']].astype(float)

x=df['HumidityAvgPercent']
y=df['DewPointAvgF']
colors=df['TempAvgF']
size=df['WindAvgMPH']*20

fig, ax=plt.subplots(figsize=(16,8))
Temp=ax.scatter(x, y, c=colors, s=size, alpha=0.7, cmap='coolwarm')
ax.set_xlabel('Humadity AVG %')
ax.set_ylabel('Dew Point AVG (F)')
ax.set_title('Austin Weather')
fig.colorbar(Temp)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 2: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
dfg=df.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].mean()
dfg.head(11)
```

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

Jawab:

```python
#code here
fig, ax=plt.subplots(figsize=(16,8))

x=np.arange(len(dfg.columns))
y=dfg.loc['Action']
y2=dfg.loc['Adventure']
y3=dfg.loc['Fighting']
y4=dfg.loc['Misc']
y5=dfg.loc['Platform']
y6=dfg.loc['Puzzle']
y7=dfg.loc['Racing']
y8=dfg.loc['Role-Playing']
y9=dfg.loc['Shooter']
y10=dfg.loc['Simulation']
y11=dfg.loc['Sports']
y12=dfg.loc['Strategy']

ax.bar(x-0.35, y, width=0.07, color='r', label='Action')
ax.bar(x-0.28, y2, width=0.07, color='b', label='Adventure')
ax.bar(x-0.21, y3, width=0.07, color='#9934FF', label='Fighting')
ax.bar(x-0.14, y4, width=0.07, color='grey', label='Misc')
ax.bar(x-0.07, y5, width=0.07, color='y', label='Platform')
ax.bar(x, y6, width=0.07, color='g', label='Puzzle')
ax.bar(x+0.07, y7, width=0.07, color='pink', label='Racing')
ax.bar(x+0.14, y8, width=0.07, color='r', label='Role-Playing')
ax.bar(x+0.21, y9, width=0.07, color='b', label='Shooter')
ax.bar(x+0.28, y11, width=0.07, color='#9934FF', label='Simulation')
ax.bar(x+0.35, y10, width=0.07, color='grey', label='Sports')
ax.bar(x+0.42, y11, width=0.07, color='y', label='Strategy')

ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
ax.set_title('Mean Sales Video Games By Genre')
ax.set_xticks(x)
ax.set_xticklabels(list(dfg.columns))
ax.legend(loc='upper right',bbox_to_anchor=(1.08,1.18))

plt.style.use('ggplot')
plt.show()
```

Dari grafik diatas dapat diketahui bahwa secara keseluruhan mean sales genre Platform menjulang tinggi yang mencapai lebih dari 0.5 pada region NA_Sales, namun dari masing-masing region yang memiliki mean sales tertinggi berasal dari Genre yang berbeda.
Yang memiliki mean sales tertinggi dari masing-masing region adalah pada region NA_Sales Genre Platform, region EU_Sales Genre Shooter, JP_Sales Genre Role-Playing, region Other_Sales Genre Shooter.

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

Jawab:

```python
#code here
fig, ax=plt.subplots(figsize=(16,8))

x=np.array(dfg.columns)
y=dfg.loc['Action']
y2=dfg.loc['Adventure']
y3=dfg.loc['Fighting']
y4=dfg.loc['Misc']
y5=dfg.loc['Platform']
y6=dfg.loc['Puzzle']
y7=dfg.loc['Racing']
y8=dfg.loc['Role-Playing']
y9=dfg.loc['Shooter']
y10=dfg.loc['Simulation']
y11=dfg.loc['Sports']
y12=dfg.loc['Strategy']

ax.bar(x, y, width=0.8, color='r', label='Action')
ax.bar(x, y2, bottom=y, width=0.8, color='b', label='Adventure')
ax.bar(x, y3, bottom=y+y2, width=0.8, color='#9934FF', label='Fighting')
ax.bar(x, y4, bottom=y+y2+y3, width=0.8, color='grey', label='Misc')
ax.bar(x, y5, bottom=y+y2+y3+y4, width=0.8, color='y', label='Platform')
ax.bar(x, y6, bottom=y+y2+y3+y4+y5, width=0.8, color='g', label='Puzzle')
ax.bar(x, y7, bottom=y+y2+y3+y4+y5+y6, width=0.8, color='pink', label='Racing')
ax.bar(x, y8, bottom=y+y2+y3+y4+y5+y6+y7, width=0.8, color='r', label='Role-Playing')
ax.bar(x, y9, bottom=y+y2+y3+y4+y5+y6+y7+y8, width=0.8, color='b', label='Shooter')
ax.bar(x, y10, bottom=y+y2+y3+y4+y5+y6+y7+y8+y9, width=0.8, color='#9934FF', label='Simulation')
ax.bar(x, y11, bottom=y+y2+y3+y4+y5+y6+y7+y8+y9+y10, width=0.8, color='grey', label='Sports')
ax.bar(x, y12, bottom=y+y2+y3+y4+y5+y6+y7+y8+y9+y10+y11, width=0.8, color='y', label='Strategy')

ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
ax.set_title('Mean Sales Video Games By Genre')
ax.legend(loc='upper right', bbox_to_anchor=(1.08, 1.2))

plt.style.use('ggplot')
plt.show()
```

Dari grafik diketahui bahwa berdasarkan kategori Genre, setiap region yaitu NA_Sales memiliki jumlah mean sales tertinggi yaitu lebih dari 2.5,sedangkan Other_Sales memiliki jumlah mean sales terendah yaitu sekitar 0.5.
